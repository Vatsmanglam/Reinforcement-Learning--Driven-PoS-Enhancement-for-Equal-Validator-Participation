```python
import time
import hashlib
import random

class Message:
    def __init__(self, data):
        self.hash = None
        self.prev_hash = None
        self.timestamp = time.time()
        self.size = len(data.encode('utf-8'))
        self.data = data
        self.payload_hash = self._hash_payload()

    def _hash_payload(self):
        return hashlib.sha256(bytearray(str(self.timestamp) + str(self.data), "utf-8")).hexdigest()

    def _hash_message(self):
        return hashlib.sha256(bytearray(str(self.prev_hash) + self.payload_hash, "utf-8")).hexdigest()

    def link(self, message):
        self.prev_hash = message.hash

    def seal(self):
        self.hash = self._hash_message()

    def validate(self):
        if self.payload_hash != self._hash_payload():
            raise InvalidMessage("Invalid payload hash in message: " + str(self))
        if self.hash != self._hash_message():
            raise InvalidMessage("Invalid message hash in message: " + str(self))

    def __repr__(self):
        return 'Message<hash: {}, prev_hash: {}, data: {}>'.format(
            self.hash, self.prev_hash, self.data
        )

class Block:
    def __init__(self, *args):
        self.messages = []
        self.timestamp = None
        self.prev_hash = None
        self.hash = None
        self.nonce = None
        self.validator = None

        if args:
            for arg in args:
                self.add_message(arg)

    def _hash_block(self):
        current_hash = hashlib.sha256(bytearray(str(self.prev_hash) + str(self.timestamp) + self.messages[-1].hash,"utf-8")).hexdigest()
        return current_hash

    def add_message(self, message):
        if len(self.messages) > 0:
            message.link(self.messages[-1])
        message.seal()
        message.validate()
        self.messages.append(message)

    def link(self, block):
        self.prev_hash = block.hash

    def seal(self, validator):
        self.timestamp = time.time()
        self.hash = self._hash_block()
        self.validator = validator

    def validate(self):
        for i, msg in enumerate(self.messages):
            try:
                msg.validate()
                if i > 0 and msg.prev_hash != self.messages[i-1].hash:
                    raise InvalidBlock("Invalid block: Message #{} has invalid message link in block: {}".format(i, str(self)))
            except InvalidMessage as ex:
                raise InvalidBlock("Invalid block: Message #{} failed validation: {}. In block: {}".format(
                    i, str(ex), str(self))
                )

    def __repr__(self):
        return 'Block<hash: {}\n, prev_hash: {}\n, validator: {}\n, messages: {}\n, timestamp: {}>\n'.format(
            self.hash, self.prev_hash, self.validator, len(self.messages), self.timestamp
        )

class BlockChain:
    def __init__(self):
        self.chain = []

    def add_block(self, block, validators):
        if len(self.chain) > 0:
            block.prev_hash = self.chain[-1].hash

        selected_validator = self.select_validator(validators)
        block.seal(selected_validator)
        block.validate()
        self.chain.append(block)

    def validate(self):
        for i, block in enumerate(self.chain):
            try:
                block.validate()
            except InvalidBlock as exc:
                raise InvalidBlockchain("Invalid blockchain at block number {} caused by: {}".format(i, str(exc)))
        return True

    def __repr__(self):
        return 'BlockChain<blocks: {}>'.format(len(self.chain))

    def select_validator(self, validators):
        total_stake = sum(v.stake for v in validators)
        rand = random.uniform(0, total_stake)
        cumulative_stake = 0
        for validator in validators:
            cumulative_stake += validator.stake
            if rand < cumulative_stake:
                return validator
        return validators[-1]

class Validator:
    def __init__(self, id, stake):
        self.id = id
        self.stake = stake

    def __repr__(self):
        return f'Validator<id: {self.id}, stake: {self.stake}>'

validators = [
    Validator(id=1, stake=10),
    Validator(id=2, stake=20),
    Validator(id=3, stake=30),
]

blockchain = BlockChain()

for i in range(3):
    messages = [Message(f"Transaction {i}-{j}") for j in range(5)]
    block = Block(*messages)

    blockchain.add_block(block, validators)
    print(blockchain.chain[-1])

if blockchain.validate():
    print("Blockchain is valid.")
else:
    print("Blockchain is invalid.")

class InvalidMessage(Exception):
    def __init__(self, *args, **kwargs):
        Exception.__init__(self, *args, **kwargs)

class InvalidBlock(Exception):
    def __init__(self, *args, **kwargs):
        Exception.__init__(self, *args, **kwargs)

class InvalidBlockchain(Exception):
    def __init__(self, *args, **kwargs):
        Exception.__init__(self, *args, **kwargs)

class InvalidNetwork(Exception):
    def __init__(self, *args, **kwargs):
        Exception.__init__(self, *args, **kwargs)
```

```python
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
import gymnasium as gym
from stable_baselines3 import DQN
from stable_baselines3.common.env_checker import check_env
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.callbacks import EvalCallback
from sklearn.model_selection import train_test_split
from stable_baselines3.common.evaluation import evaluate_policy

data = pd.read_csv("/content/updated_dataset3.csv")
data = data.dropna()

features = ['number', 'difficulty', 'total_difficulty', 'size', 'gas_limit', 'gas_used', 'transaction_count']
data = data.drop(columns=['hash', 'parent_hash', 'nonce', 'sha3_uncles', 'logs_bloom', 'transactions_root',
            'state_root', 'receipts_root', 'extra_data', 'Age of the block', 'Status of transaction'])

scaler = MinMaxScaler()
posClass = Blockchain()
data[features] = scaler.fit_transform(data[features])

train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

class BlockchainEnv(gym.Env):
    def __init__(self, df):
        super(BlockchainEnv, self).__init__()
        self.df = df.copy()
        self.current_step = 0
        self.miner_addresses = self.df['miner'].unique()
        self.action_space = gym.spaces.Discrete(len(self.miner_addresses))
        self.observation_space = gym.spaces.Box(low=0, high=1, shape=(len(features),), dtype=np.float32)

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.current_step = 0
        observation = self._next_observation()
        info = {}
        return np.array(observation), info

    def step(self, action):
        if self.current_step < len(self.df):
            true_miner_index = self.df.iloc[self.current_step]['miner']
            reward = 1 if action == true_miner_index else 0
            self.current_step += 1
        else:
            reward = 0
        terminated = self.current_step >= len(self.df)
        truncated = False

        observation = self._next_observation()
        info = {}
        return observation, reward, terminated, truncated, info

    def _next_observation(self):
        if self.current_step >= len(self.df):
            return np.zeros(len(features), dtype=np.float32)
        obs = np.array(self.df.iloc[self.current_step][features].values, dtype=np.float32)
        return obs

    def render(self, mode='human'):
        pass

train_env = BlockchainEnv(train_data)
test_env = BlockchainEnv(test_data)
check_env(train_env, warn=True)

eval_callback = EvalCallback(test_env, best_model_save_path='./logs/',
                             log_path='./logs/', eval_freq=500,
                             deterministic=True, render=False)

iteration_data = []
time_data = []
cumulative_time_data = []

total_timesteps = 50000
start_time = time.time()
cumulative_time = 0

model = DQN('MlpPolicy', train_env, verbose=1, learning_rate=0.0001,
            buffer_size=50000, exploration_fraction=0.1, exploration_final_eps=0.02)

for iteration in range(1, total_timesteps + 1):
    model.learn(total_timesteps=1, callback=eval_callback)

    iteration_time = time.time() - start_time
    iteration_data.append(iteration)
    time_data.append(iteration_time)

    cumulative_time += iteration_time
    cumulative_time_data.append(cumulative_time)

    start_time = time.time()

    # print(f"Iteration: {iteration}, Time taken: {iteration_time:.2f} seconds")

plt.figure(figsize=(10, 6))
plt.plot(iteration_data, time_data, marker='o', linestyle='-')
plt.title('Time taken per iteration')
plt.xlabel('Iteration')
plt.ylabel('Time (seconds)')
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(iteration_data, cumulative_time_data, marker='o', linestyle='-')
plt.title('Learning Curver vs Training Iterations')
plt.xlabel('Iteration')
plt.ylabel('Learning Curve (Mean Reward)')
plt.grid(True)
plt.tight_layout()
plt.show()

training_rewards = []
for i in range(5000):
    model.learn(total_timesteps=1)
    training_reward, _ = evaluate_policy(model, train_env, n_eval_episodes=10)
    training_rewards.append(training_reward)

plt.figure(figsize=(10, 6))
plt.plot(training_rewards, marker='o', linestyle='-')
plt.xlabel('Iteration')
plt.ylabel('Mean Reward')
plt.title('Mean Reward vs. Training Iterations')
plt.grid(True)
plt.tight_layout()
plt.show()

for _ in range(len(test_data)):
    obs, _ = test_env.reset()
    done = False
    while not done:
        action, _states = model.predict(obs, deterministic=True)
        obs, reward, done, _, _ = test_env.step(action)
    if not done:
        predicted_miner_index = action[0]
        predicted_miner_address = test_data.iloc[predicted_miner_index]['miner']
        print("Predicted Miner Address:", predicted_miner_address)



```
